```python
import numpy as np
import matplotlib.pyplot as plt
```

**Question 6**

```python
N = 100
```

```python
def NgroundOverN_2states_MB (kBToverEpsilon) :
    betaE = 1/kBToverEpsilon
    return 1 / (1+np.exp(- betaE))

def NgroundOverN_2states_BE (kBToverEpsilon) :
    betaE = 1/kBToverEpsilon
    return (1/N)* 1 / (np.exp(- betaE)-1) + (N+1)/N * (1 / (1 - np.exp(- (N+1) * betaE)))
```

`x` est le vecteur qui contient les valeurs de $k_B T / \varepsilon_1$

```python
x = np.linspace(0.1, 200, 200)
```

```python
fig, ax = plt.subplots()

plt.plot(x, NgroundOverN_2states_MB(x), label='Maxwell-Boltzmann')
plt.plot(x, NgroundOverN_2states_BE(x), label='Bose-Einstein')

ax.set_ylim([0, 1])
ax.set_xlabel(r'$\frac{1}{\beta \epsilon}$')
ax.set_ylabel(r'$\frac{\langle N_0 \rangle}{N}$')
ax.legend()
```

**Question 12**

```python
term_MB = lambda nx, ny, nz, betaE, Ng : Ng*np.exp(-betaE*(nx*nx+ny*ny+nz*nz))
term_BE = lambda nx, ny, nz, betaE, Ng : np.exp(-betaE*(nx*nx+ny*ny+nz*nz)) / ((1+1/Ng) - np.exp(-betaE*(nx*nx+ny*ny+nz*nz)))

Nmax = 20 # max value for nx, ny or nz

def N_MB(kBToverEpsilon, Ng):
    betaE = 1/kBToverEpsilon
    nx, ny, nz = np.arange(Nmax), np.arange(Nmax), np.arange(Nmax)
    out = 0
    for i in nx:
        for j in ny:
            for k in nz:
                out += term_MB(i, j, k, betaE, Ng)
    return out

def N_BE(kBToverEpsilon, Ng):
    betaE = 1/kBToverEpsilon
    nx, ny, nz = np.arange(Nmax), np.arange(Nmax), np.arange(Nmax)
    out = 0
    for i in nx:
        for j in ny:
            for k in nz:
                out += term_BE(i, j, k, betaE, Ng)
    return out

def NgroundOverN_MB_exact(kBToverEpsilon):
    betaE = 1/kBToverEpsilon
    return (4 * betaE/np .pi)**(1.5)
```

```python
from scipy.optimize import fsolve

def cost_MB(Nground, *args):
    """
    find Nground for a given N depending on kBToverEpsilon for MB statistics
    """
    kBToverEpsilon, N = args
    return N - N_MB(kBToverEpsilon, Nground)

def cost_BE(Nground, *args):
    """
    find Nground for a given N depending on kBToverEpsilon for BE statistics
    """
    kBToverEpsilon, N = args
    return N - N_BE(kBToverEpsilon, Nground)
```

```python
num = 100                      # number of points for kB T / Epsilon
x = np.linspace(1., 40., num)  # values for [ kB T / Epsilon ]
N = 100                        # number of particles
```

```python
NgroundOverN_MB = np.array([])
Ng = N # initial guess for Nground (very close to N at T ~ 0)

for kTE in x:
    args = (kTE, N)
    root = fsolve(cost_MB, Ng, args, xtol = 0.01)
    Ng = root[0]
    # print("kTE = {0:5.2f}  ---  Ng = {1:6.2f}".format(kTE, root[0]))
    NgroundOverN_MB = np.concatenate([NgroundOverN_MB, [Ng/N]])
```

```python
NgroundOverN_BE = np.array([])
Ng = N # initial guess for Nground (very close to N at T ~ 0)

for kTE in x:
    args = (kTE, N)
    root = fsolve(cost_BE, Ng, args, xtol = 0.01)
    Ng = root[0]
    # print("kTE = {0:5.2f}  ---  Ng = {1:6.2f}".format(kTE, root[0]))
    NgroundOverN_BE = np.concatenate([NgroundOverN_BE, [Ng/N]])
```

```python
fig, ax = plt.subplots()

plt.plot(x, NgroundOverN_MB, label='Maxwell-Boltzmann')
plt.plot(x, NgroundOverN_BE, label='Bose-Einstein')
plt.plot(x, NgroundOverN_MB_exact(x), linestyle='dotted', label='Approximation MB')

ax.set_xlim([0, 40])
ax.set_ylim([0, 1])
ax.set_xlabel(r'$\frac{1}{\beta \epsilon}$')
ax.set_ylabel(r'$\frac{\langle N_0 \rangle}{N}$')
ax.legend()
```

**Question 17**

```python
NgroundOverN_MB_approx = lambda ToverTC : 1-ToverTC**(1.5)

y = np.linspace(0.1, 1, 100)

def ToverTC_(kBToverEpsilon) :
    return (0.25*np.pi*kBToverEpsilon)*(2.612/N)**(0.666)

def betaE_(ToverTC, N) :
    return (0.25*np.pi/ToverTC)*(2.612/N)**(0.666)

def kBToverEpsilon_(betaE) :
    return 1/betaE

z = kBToverEpsilon_(betaE_(x, N))
z = ToverTC_(x)
```

```python
fig, ax = plt.subplots()

plt.plot(z, NgroundOverN_BE, label='Bose-Einstein')
plt.plot(y, NgroundOverN_MB_approx(y), 'k', linestyle='dotted', label = 'Analytic')
plt.plot([1, 3], [0, 0], 'k', linestyle='dotted')

ax.set_xlim([0, 3])
ax.set_ylim([-0.1, 1])
ax.set_xlabel(r'$T/T_C$')
ax.set_ylabel(r'$\frac{\langle N_0 \rangle}{N}$')
ax.legend()
```
