```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
```

Définition du couple de paramètres $a$ et $b$ pour notre gaz de van der Waals.

**Note** : aucune grandeur n'est physique, et on ne s'embête pas trop avec les valeurs de différentes constantes. L'objectif de ce notebook est de servir une discussion qualitative du phénomène de la transition de phase.

```python
ab = (1e-4, 1e-4)
```

Valeurs de références de $T$, $p$ et $N$ : les valeurs ne sont donc toujours pas physiques. On les emballent dans un tuple pour les protéger.

```python
TpN = (0.288, 350, 1)
```

```python
def triple(TpN, ab):
    """point triple du gaz de van der Waals, fonction de a et b (mais aussi de N)"""
    a, b = ab
    N = TpN[2]
    return {'Tc':8*a/(27*b), 'Vc':3*b*N, 'pc':a/(27*b*b)}
```

```python
triple(TpN, ab)
```

```python
def p_func(V, T, N, ab):
    """p(T, V, N) pour un gaz de van der Waals"""
    a, b = ab
    return N*T/(V-b*N) - a*(N/V)**2
```

```python
V = np.linspace(0.0002, 0.0005, 100) # un span pour les volumes... définis autour du volume criqique
```

```python
T, _, N = TpN

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(V, p_func(V, T, N, ab))

ax.set_xlabel('$V$')
ax.set_ylabel('$p$')
ax.set_title('$p(V)$ pour un gaz de van der Waals à $T < T_c$')
```

```python
def p_prime(V, T, N, ab):
    """derivee de p(V) pour un gaz de van der Waals"""
    a, b = ab
    return - N*T/(V-b*N)**2 + 2*a*N**2/V**3
```

```python
from scipy.optimize import fsolve

Va = fsolve(p_prime, x0=0.00025, args=(T, N, ab))
Vb = fsolve(p_prime, x0=0.00035, args=(T, N, ab))

pa = p_func(Va, T, N, ab)
pb = p_func(Vb, T, N, ab)
```

```python
T, _, N = TpN

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(V, p_func(V, T, N, ab))

ax.plot(Va, pa, marker='o')
ax.plot(Vb, pb, marker='o')

ax.text(Va, 1.01*pa, "a")
ax.text(Vb, 1.01*pb, "b")

ax.set_xlabel('$V$')
ax.set_ylabel('$p$')
ax.set_title('$p(V)$ pour un gaz de van der Waals à $T < T_c$')
```

Les points oranges et verts définissent les pressions $p_a$ et $p_b$ donnant les min et max de $p(V)$ pour un gaz de van der Waals, à la température donnée.

```python
def deBroglie(TpN, m=1):
    """longueur d'onde thermique"""
    T, p, N = TpN
    return 1/np.sqrt(2*np.pi*m*T)
```

```python
def F(TpN, V, ab=ab):
    """energie libre pour un gaz de van der Waals"""
    T, _, N = TpN
    a, b = ab
    beta = 1/T
    Lambda = deBroglie(TpN)
    return N*T*(np.log(N/(V-b*N))+3*np.log(Lambda)-1-beta*a*(N/V))
```

```python
def Phi(TpN, V):
    T, p, N = TpN
    return F(TpN, V)+p*V
```

```python
T, _, N = TpN
p = 328
TpN = [T, p, N]

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(V, Phi(TpN, V))
ax.set_xlabel('$V$')
ax.set_ylabel('$F(T, V, N)+pV$')
ax.set_title('$p=328 < p_{Maxwell}$ pour un gaz de van der Waals')
```

Ce premier plot est à basse pression, ie pour une pression supérieure à la première pression pour laquelle $p^{\prime}(V) = 0$. Le puit le plus profond est le second, à grand volume $V$. L'état le plus stable est donc celui à grand volume $V$ (associé au deuxième puit), pour lequel le système est alors gazeux. Mais si la pression n'est pas trop faible, il est aussi possible d'y observer la phase liquide qui est alors métastable.

```python
T, _, N = TpN
p = 334
TpN = [T, p, N]

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(V, Phi(TpN, V))
ax.set_xlabel('$V$')
ax.set_ylabel('$F(T, V, N)+pV$')
ax.set_title('$p=334 > p_{Maxwell}$ pour un gaz de van der Waals')
```

Ce second plot est à plus haute pression, ie pour une pression supérieure à la seconde pression pour laquelle $p^{\prime}(V) = 0$. Le puit le plus profond est le premier, à faible volume $V$. L'état le plus stable est donc celui à faible volume $V$ (associé au premier puit), pour lequel le système est alors liquide. Mais si la pression n'est pas trop grande, il est aussi possible d'y observer la phase gazeuse qui est alors métastable.

```python
T, _, N = TpN
p = 330
TpN = [T, p, N]

fig, ax = plt.subplots(figsize=(6,4))

ax.plot(V, Phi(TpN, V))
ax.set_xlabel('$V$')
ax.set_ylabel('$F(T, V, N)+pV$')
ax.set_title('$p=330 \\sim p_{Maxwell}$ pour un gaz de van der Waals')
```

Avec un peu plus de courage... il faudrait calculer la forme analytique de $\int _{\alpha}^{\beta} p(V) \mathrm d V$ pour un gaz de van der Waals, puis chercher dans l'interval $[p_a, p_b]$ la pression $p_c$ (associée à $V_c$) telle que $\int_{V_a}^{V_c} p(V) \mathrm d V = \int_{V_c}^{V_b} p(V) \mathrm d V$. Cette pression $p_c$ définit la valeur de la pression à laquelle le changement de phase se fait... suivant la loi du palier de Maxwell.
